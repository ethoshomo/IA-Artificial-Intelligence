```python
import matplotlib.pyplot as plt
```

A ideia principal do matplotlib é reunir uma coleção de funções que fazem modificações em uma figura. Isto é, a figura será criada, será criada uma área de plotagem, linhas de plotagem e todas as decorações com labels e outros aspectos inerentes à construção de um gráfico.

> Axes: não é utilizado na acepção matemática, sendo comumente utilizado para se referir a parte da imagem (figure).

## PLOT()

O plot é usado para criar o gráfico. 

Por default, a função atribui aos dados um valor indefinido (None), bem como cria automaticamente a escala do eixo X (scalex=True) e do eixo Y (scaley=). É de se destacar também que recebe **kwargs (é interessante criar uma variável a parte para passar todos os demais parâmetros `**<variavel>`).

* Se passarmos apenas um único parâmetro, a função irá subentender que se trata do array de valores de Y e irá subentender uma série iniciada em 0 (zero) para o X.

* Se passarmos dois argumentos, a função vai entender que se trata de um array X e de um array Y, de modo que, combinados, formaram pares de pontos que serão plotados. Observação: os arrays precisam necessariamente ser do mesmo tamanho (caso contrário, vai dar erro).

* É possível passarmos uma lista de argumentos sobre os marcadores e linhas (color=, marker=, linestyle=, linewidth=, markersize=). Existe uma forma de condensar esses valores usando a formatação de strings `ftm='[marker][line][color]'`

* O uso do plt.show() até acontece no Jupiter, porém é recomendado seu uso.

```python
x=[1,2,3,4,5]
y=[2,7,4,3,10]

plt.plot(x, y)
plt.show()
```

## Elementos que podem ser alterados

* **_Figure_**: trata-se do container criado por uma classe que irá implementar a figura base, a qual receberá todos os elementos plotados pelo Matplotlib. Existem vários parâmetros que podemos manipular: figsize=, dpi=, facecolor=, edgecolor=, linewidth=, etc.

> É possível adicionar SubFigures (Figure.addsubfigure). Essa funcionalidade tem diversos objetivos, mas o mais interessante é a criação de um grid.

> É possível salvar a figura criada por meio do plt.savefig(), que recebe como parâmetro a string com o nome e a extensão da imagem ('fig.png').

> O uso do subplot é interessante e merece atenção. Exemplo: `<figure>.add_suplot(<number>)`, sendo que a number é um número de tres dígitos. O primeiro indica a quantidade de linhas, o segundo a quantidade de colunas por linha e o terceiro exatamente o numero do gráfico. Isto é, 131 informa que se trata de uma figura que tem uma linha 3 colunas e se trata do primeiro subplot.

* **_Textos_**: existem alguns textos que podem ser inseridos no gráfico, tais como Títulos, Nomes dos Eixos X e Y, bem como textos de legendas.

    * **_Title_**: trata-se do método que adiciona um título ao gráfico.

    * **_Label_**: trata-se de um parâmetro da função do método plot(), que nomeia a marcação de plotagem feita no gráfico. Isto é, se for uma linha, estaremos nomeando o conjunto de pontos gerados, sendo certo que este label será usado, por exemplo, na legenda.

    * **_X Axis Label_**: trata-se de método próprio usado para indicar o label do eixo X. Como parâmetro, recebe a string que será plotada.

    * **_Y Axis Label_**: trata-se de método próprio usado para indicar o label do eixo Y. Como parâmetro, recebe a string que será plotada.

* **_Ticks_**: ticks são as escalas que aparecem nos gráficos. Nesse caso, temos as escalas maiores (por exemplo, valores inteiros) e as menores (por exemplo, valores decimais dos inteiros). 

    * **_Major Tick_**: recebem como parâmetro um array de valores que irão definir a escala do gráfico. Não é necessário indicar todos os valores que farão a composição dos dados, pois o próprio programa irá calcular a escala inicial e final e irá plotar os valores de acordo com ela. Porém, omitir um determinado valor do array irá deixar de mostrar o seu valor no eixo de referência. Sintaxe: `plt.xticks([1,2,3])`

    * **_Minor Tick_**: trata-se da subescala do gráfico. É opcional seu uso.

    * **_Major Tick Label e Minor Tick Label_**: é possível fazer alterações nos valores dos ticks, mas convém olhar a documentação. Talvez nem valha a pena.

* **_Markers_**: trata-se dos marcadores de pontos que são plotados. Podem ser pontos ('.'), pixel (','), círculos ('o'), triângulos ('v', '^'), etc.

* **_Line_**: trata-se da forma como as linhas plotadas assumem. Podemos indicar os seguintes estilos: a) linha sólida ('-'); b) dashed ('--'); c) dashed-dot ('-.'); d) dotted line (':').

* **_Colors_**: nos formatos de string, podemos indicar as cores de forma abreviada: a) blue ('b'); b) green ('g'); c) red ('r'); etc.

* **_Axes_**: são quadrantes do gráfico.

* **_Legend_**: podemos adicionar legenda no gráfico

```python
plt.plot(x,y, label='Teste')
plt.title('Título do Gráfico')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.tick_params()
plt.xticks([-1,2,3,4,5,6,7])
plt.yticks([-1,2,3,4,5,6,7])

plt.show()
```

```python
valoresx = [1, 2, 3, 4]
valoresy = [1, 4, 2, 3]


figura = plt.figure(figsize=(20,5))
figura.suptitle('Título Geral')


figura.add_subplot(131)
plt.plot(valoresx, valoresy, label='Um dado qualquer')
plt.legend()
plt.title('Gráfico 1')

figura.add_subplot(132)
plt.scatter(valoresx,valoresy)
plt.title('Gráfico 2')

figura.add_subplot(133)
plt.bar(valoresx, valoresy)
plt.title('Gráfico 3')

#plt.savefig('gráficos.png')

plt.show(figura)
```

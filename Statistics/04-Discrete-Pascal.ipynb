# Pascal Distribution

Pascal's Distribution is another Bernoulli sequence of events. It is used when we do not know the number of events that will be necessary, but we can calculate the probability of a given expectation. For example, we could answer what is the chance of a basketball player making the fifth basket on the tenth attempt.

```python
p = 0.6 # P('Success')
n = 10 # total number of events
r = 5 # total successes in 'n' events
```

### Probabily Mass Function (PMF)

$$B(n; k, p) = \binom{n-1}{k-1}p^{k}(1-p)^{n-k}$$

Where:

* n = total experiments
* k = number of successes
* p = probability of success

```python
def pascal_pmf(p, n, k):
    """ 
    p: percentage off success
    n: total of experiments
    k: number of successful experiments, with the 
        last launch being a success.
    """  
    numerator = math.factorial(n-1)
    denominator = math.factorial(k-1)*math.factorial((n-1)-(k-1))
    
    p0 = numerator/denominator
    p1 = p**k
    p2 = (1-p)**(n-k)
    return p0*p1*p2
```

```python
pascal_x = []
pascal_y = []

for k in range(r, n+1):
    pascal_x.append(k)
    pascal_y.append(pascal_pmf(p, n, k))

print(len(pascal_x), pascal_x)
print(len(pascal_y),pascal_y)
```

### Probability Mass Function Chart

```python
plt.figure(figsize=(4,4))
plt.bar(pascal_x,pascal_y)
plt.title("Pascal - Probability Mass Function Chart")
plt.xlabel("Attemps with the last one success")
plt.ylabel("Probability")
plt.show()
```


### Cumulative Distribution Function (CDF)

```python
before = 0
pascal_y_cumulative = []
for number in pascal_y:
    before = before + number
    pascal_y_cumulative.append(before)

plt.figure(figsize=(4,4))
plt.bar(pascal_x, pascal_y_cumulative)
plt.title("Pascal - Probability Mass Function Chart")
plt.xlabel("Results")
plt.ylabel("Probability")
plt.show()
```

### Expectation (E(X))

$$E(x) = \frac{r}{p}$$

```python
def pascal_expectation(r, p):
    return r/p
```

```python
p_expectation = pascal_expectation(r, p)
print(p_expectation)
```

### Variance (Var(x))

$$Var(X) = \frac{r(1-p)}{p^{2}}$$

```python
def pascal_variance(r, p):
    return r*(1-p)/p**2
```

```python
print(pascal_variance(r, p))
```


### Standard Deviation (STD(X))

```python
def pascal_std(r, p):
    return pascal_variance(r, p)**(1/2)
```

```python
print(pascal_std(r, p))
```

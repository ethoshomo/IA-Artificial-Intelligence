# Geometric Distribution


The Geometric Distribution is a discrete probability distribution that models the number of independent attempts to obtain the first success in a Bernoulli process. However, this process can be seen from the perspective of failures, as we can approach the same topic by looking at how many failures were necessary to obtain the first success.

### Probabily Mass Function (PMF)


$$P(X = k) = (1-p)^{k-1}*p$$

```python
n = 10 # total attempts (including success)
p = 0.4 # probability of success
```

```python
def geometric_pmf(p, k):
    return p*((1-p)**(k-1))
```

```python
geometric_x = [x for x in range(1,n)]
geometric_y = [geometric_pmf(p, k) for k in geometric_x]

print(geometric_x)
print(geometric_y)
```

```python
plt.figure(figsize=(4,4))
plt.bar(geometric_x, geometric_y)
plt.title("Geometric - Probability Mass Function Chart")
plt.xlabel("Total Attempts (n)")
plt.ylabel("Probability")
plt.show()
```

### Cumulative Distribution Function (CDF)


```python
cumulative = 0
geometric_cumulative = []
for v in geometric_y:
    cumulative = cumulative + v
    geometric_cumulative.append(cumulative)
```

```python
plt.figure(figsize=(4,4))
plt.bar(geometric_x, geometric_cumulative)
plt.title("Geometric - Probability Mass Function Chart")
plt.xlabel("Total Attempts (n)")
plt.ylabel("Probability")
plt.show()
```

### Expectation (E(X))


$$E(x) = \frac{1}{p}$$

```python
def geometric_expecatation(p):
    return 1/p
```

```python
scipy_expectation = geom.mean(p, loc=0)
hard_coded_expectation = geometric_expecatation(p)

print(scipy_expectation, hard_coded_expectation)
```

### Variance (Var(x))


$$Var(X) = \frac{1-p}{p^{2}}$$

```python
def geometric_variance(p):
    return (1-p)/p**2
```

```python
scipy_variance = geom.var(p, loc=0)
hard_coded_variance = geometric_variance(p)

print(scipy_variance, hard_coded_variance)
```

### Standard Deviation (STD(X))

```python
def geometric_std(p):
    return geometric_variance(p)**(1/2)
```

```python
scipy_std = geom.std(p, loc=0)
hard_coded_std = geometric_std(p)

print(scipy_std, hard_coded_std)
```

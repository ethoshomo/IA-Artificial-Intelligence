# Poisson Distribution

Although it is a discrete distribution, the Poisson distribution does not use the binomial distribution like others. It expresses the probability of a number of events occurring in a fixed interval of time or space as long as the average rate of occurrence since the last event is known.

### Probabily Mass Function (PMF)


$$P(X=k) = \frac{e^{-\lambda}\lambda^{k}}{k!}$$

Where: 

* e: natural logarithm
* k: 0, 1, 2, ...
* $\lambda$: average rate per unit of time

```python
lambda_ = 8
k = 20
```

```python
def poisson_pmf(lambda_, k):
    return (math.e**(-lambda_)*lambda_**k)/math.factorial(k)
```

```python
poisson_x = []
poisson_y = []
for n in range(k):
    poisson_x.append(n)
    poisson_y.append(poisson_pmf(lambda_, n))
```

```python
plt.figure(figsize=(4,4))
plt.bar(poisson_x, poisson_y)
plt.title("Poisson - Probability Mass Function Chart")
plt.xlabel("Total Attempts (n)")
plt.ylabel("Probability")
plt.show()
```

### Cumulative Distribution Function (CDF)


```python
cumulative = 0
poisson_cumulative = []
for v in poisson_y:
    cumulative = cumulative + v
    poisson_cumulative.append(cumulative)
```

```python
plt.figure(figsize=(4,4))
plt.bar(poisson_x, poisson_cumulative)
plt.title("Poisson - Probability Mass Function Chart")
plt.xlabel("Total Attempts (n)")
plt.ylabel("Probability")
plt.show()
```

### Expectation (E(X))


$$E(X) = \lambda$$

```python
poisson.mean(lambda_)
```

### Variance (Var(x))


$$Var(X) = \lambda$$

```python
poisson.var(lambda_)
```

### Standard Deviation (STD(X))


```python
def poisson_std(lambda_):
    return lambda_**(1/2)
```

```python
print(poisson.std(lambda_), poisson_std(lambda_))
```
